```python
# Scatter plot
fig, ax = plt.subplots(figsize = (18,10))
ax.scatter(df['math score'], df['reading score'])

# x-axis label
ax.set_xlabel('math score')

# y-axis label
ax.set_ylabel('reading score')
plt.show()
```

```python
# Position of the Outlier
print(np.where((df['math score']<20)))
```

```python
# Z score
from scipy import stats
import numpy as np

z = np.abs(stats.zscore(df['math score']))
print(z)
```

```python
threshold = 3

# Position of the outlier
print(np.where(z > 3))
```

```python
# IQR
Q1 = np.percentile(df['math score'], 25,
				interpolation = 'midpoint')

Q3 = np.percentile(df['math score'], 75,
				interpolation = 'midpoint')
IQR = Q3 - Q1
print(Q3," ",Q1)
print(IQR)
```

```python
# Above Upper bound
upper=df['math score'] >= (Q3+1.5*IQR)
print("Upper bound:",(Q3+1.5*IQR))
print(np.where(upper))

# Below Lower bound
lower = df['math score'] <= (Q1-1.5*IQR)
print("Lower bound:", (Q1-1.5*IQR))
print(np.where(lower))
```
